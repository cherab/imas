# Bolometer

Here we demonstrate how to load bolometer observers from a bolometer IDS, and visualize them to check what they look like.
The dataset used in this example is a mock bolometer dataset, which consists of different types of bolometer cameras.

```python
from plotly import io

from cherab.imas.datasets import bolometer_moc
from cherab.imas.observer.bolometer import load_bolometers, visualize

io.renderers.default = "notebook"
```

## Load bolometer dataset


```python
path = bolometer_moc()
bolometers = load_bolometers(path, "r")
```

```python
for bolo in bolometers:
    print(f"Bolometer: {bolo.name}")
```

## Visualize bolometer observers

### Pinhole camera

```python
fig = visualize(
    bolometers[0],
    num_rays=100,
    ray_terminate_distance=2.0e-2,
    show=False,
)
fig.update_layout(template="plotly_dark")
```

### Collimator cameras

```python
fig = visualize(
    bolometers[1],
    num_rays=100,
    ray_from_channel=[0, 3],
    ray_terminate_distance=2.0e-2,
    show=False,
)
fig.update_layout(template="plotly_dark")
```

```python
fig = visualize(
    bolometers[2],
    num_rays=200,
    ray_from_channel=[0, 3],
    ray_terminate_distance=2.0e-2,
    show=False,
)
fig.update_layout(template="plotly_dark")
```
